```python
%matplotlib inline
import matplotlib.pyplot as plt
from pycalphad import Database, binplot
import pycalphad.variables as v
db_alni = Database('alni-espei.tdb')
fig = plt.figure(figsize=(9,6))
binplot(db_alni, ['AL', 'NI', 'VA'] , list(db_alni.phases.keys()), {v.X('AL'):(1e-5,1,0.05),
                                                       v.T: (300, 2000, 30), v.P:101325}, ax=plt.gca())
plt.show()
```
